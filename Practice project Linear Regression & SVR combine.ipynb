```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(realestate.drop(['No','Y house price of unit area'],axis=1),realestate['Y house price of unit area'],test_size=0.30,random_state=20)
```

```python
model_LR=LinearRegression()
model_LR.fit(x_train,y_train)
```

```python
accuracy_LR=model_LR.score(x_test,y_test)
print(accuracy_LR*100,'%')
```

# By SVR

```python
from sklearn.metrics import accuracy_score
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(realestate.drop(['No','Y house price of unit area'],axis=1),realestate['Y house price of unit area'],test_size=0.20,random_state=20)
```

```python
model_svr=SVR()
model_svr.fit(x_train,y_train)
```

```python
accuracy_svr=model_svr.score(x_test,y_test)
print('Testing accuracy :',accuracy_svr*100,'%')
```

```python
accuracy_svr=model_svr.score(x_train,y_train)
print('Traning accuracy :',accuracy_svr*100,'%')
```
